## Part (a) Fitting a Histogram

Let $\theta = (\theta_0, \dots, \theta_{d-1}) \in \mathbb{R}^d$ and define the model $p_\theta(x) = \frac{e^{\theta_x}}{\sum_{x'}e^{\theta_{x'}}}$

Fit $p_\theta$ with maximum likelihood via stochastic gradient descent on the training set, using $\theta$ initialized to zero. Use your favorite version of stochastic gradient descent, and optimize your hyperparameters on a validation set of your choice.

**You will provide these deliverables**


1.   Over the course of training, record the average negative log-likelihood (nats / dim) of the training data (per minibatch) and test data (for your entire test set). Code is provided that automatically plots the training curves. 
2.   Report the final test set performance of your final model
3. Plot the model probabilities in a bar graph with $\{0,\dots,d-1\}$ on the x-axis and a real number in $[0,1]$ on the y-axis.




```python
import torch
import torch.nn as nn
import numpy as np
import torch.nn.functional as F
from torch.utils.data import DataLoader
class Model(nn.Module):
    def __init__(self, d):
        super().__init__()
        self.d = d
        self.theta = nn.Parameter(torch.zeros(self.d))
        
    def forward(self):
        return F.log_softmax(self.theta)
    
    def probs(self):
        return F.softmax(self.theta).detach().cpu().numpy()

def train(model, optimizer, criterion, train_loader):
    model.train()
    loss_ = []
    for data_ in train_loader:
        optimizer.zero_grad()
        out = model()
        out = out.view(1, out.shape[0])
        out = torch.cat(data_.shape[0]* [out], dim=0)
        loss = criterion(out, data_)
        loss_.append(loss.item())
        loss.backward()
        optimizer.step()
    return loss_

def eval(model, criterion, test_loader):
    model.eval()
    loss_ = 0
    for data_ in test_loader:
        out = model()
        out = out.view(1, out.shape[0])
        loss = criterion(out, data_)
        loss_ += loss.item()
    return loss_/len(test_loader)

def q1_a(train_data, test_data, d, dset_id):
    train_loader = DataLoader(train_data, batch_size=48, shuffle=True)
    test_loader = DataLoader(test_data)
    model = Model(d)
    criterion = nn.NLLLoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=0.3)
    scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.1)
    tr_loss, te_loss = [], []
    te_loss.append(eval(model, criterion, test_loader))
    for i in range(15):
        tr_loss.extend(train(model, optimizer, criterion, train_loader))
        te_loss.append(eval(model, criterion, test_loader))
        scheduler.step()
    probs = model.probs()
    return tr_loss, te_loss, probs
q1_save_results(1, 'a', q1_a)
```

## Part (b) Fitting Discretized Mixture of Logistics

Let us model $p_\theta(x)$ as a **discretized** mixture of 4 logistics such that $p_\theta(x) = \sum_{i=1}^4 \pi_i[\sigma((x+0.5 - \mu_i)/s_i) - \sigma((x-0.5-\mu_i)/s_i)]$

For the edge case of when $x = 0$, we replace $x-0.5$ by $-\infty$, and for $x = d-1$, we replace $x+0.5$ by $\infty$.

You may find the [PixelCNN++](https://arxiv.org/abs/1701.05517) helpful for more information on discretized mixture of logistics.

**Provide the same set of corresponding deliverables as part (a)**

```python
#model fails without exp->log, without clamp. 
class Model(nn.Module):
    def __init__(self, d, mix=4):
        super().__init__()
        self.d = d
        self.mix = mix
        self.theta = nn.Parameter(torch.randn(self.mix))
        self.mean_ = nn.Parameter(torch.arange(self.mix).float()/(self.mix - 1)*d)
        self.scale_ = nn.Parameter(torch.randn(self.mix))

    def forward(self, x):
        mean, scale = self.mean_.unsqueeze(0), self.scale_.unsqueeze(0)
        x = torch.cat(self.mix*[x.float().view(1,x.shape[0])]).T
        scale = torch.exp(-scale)
        bounded = torch.sigmoid((x + 0.5 - mean)*scale) - torch.sigmoid((x - 0.5 - mean)*scale)
        edge0 = torch.sigmoid((0.5 - mean)*scale)
        edge99 =1 - torch.sigmoid((self.d - 1.5 - mean)*scale)
        fn = torch.where(x<0.001, edge0, torch.where(x>self.d - 1 - 1e-3, edge99, bounded))
        log_fn = torch.log(torch.clamp(fn, min=1e-12))
        theta_fn = F.log_softmax(self.theta, dim=0).unsqueeze(0)
        return torch.logsumexp(log_fn+theta_fn, dim=1)

    def probs(self):
        val = torch.FloatTensor(torch.arange(self.d))
        return self(val).exp().detach().cpu().numpy()

def train(model, optimizer, train_loader):
    model.train()
    loss_ = []
    for data_ in train_loader:
        optimizer.zero_grad()
        loss = model(data_)
        loss = -torch.mean(loss)
        loss_.append(loss.item())
        loss.backward()
        optimizer.step()
    return loss_

def eval(model, test_loader):
    model.eval()
    loss_ = 0
    for data_ in test_loader:
        loss = model(data_)
        loss = -torch.mean(loss)
        loss_ += loss.item()
    return loss_/len(test_loader)

def q1_b(train_data, test_data, d, dset_id):
    train_loader = DataLoader(train_data, batch_size=20, shuffle=True)
    test_loader = DataLoader(test_data)
    d = float(d)
    model = Model(d)
    optimizer = torch.optim.Adam(model.parameters(), lr=0.1)
    scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=2, gamma=0.1)
    tr_loss, te_loss = [], []
    te_loss.append(eval(model, test_loader))
    for i in range(5):
        tr_loss.extend(train(model, optimizer, train_loader))
        te_loss.append(eval(model, test_loader))
        scheduler.step()
    probs = model.probs()
    return tr_loss, te_loss, probs
q1_save_results(1, 'a', q1_b)
```

Implement and train a MADE model through maximum likelihood to represent $p(x_0, x_1)$ on the given datasets, with any autoregressive ordering of your choosing. 

A few notes:

* You made find it useful to one-hot encode your inputs. 

**You will provide these deliverables**

3. Visualize the learned 2D distribution by plotting a 2D heatmap


```python
#keras to_categorical fails on some batches(same number of input), use explicit one hot. mask code referred from Karpathy's MADE implementation, 
#grid code taken from original solution
from keras.utils import to_categorical
import torch
import torch.nn as nn
import numpy as np
import torch.nn.functional as F
from torch.utils.data import DataLoader

class mask_operation(nn.Linear):
    def __init__(self, in_features, out_features, bias=True):
        super().__init__(in_features, out_features, bias)
        self.register_buffer('mask', torch.ones(out_features, in_features))
    
    def forward(self, x):
        return nn.functional.linear(x, self.mask*self.weight, self.bias)
  
    def create(self, mask):
        self.mask.data.copy_(torch.from_numpy(mask.astype(np.uint8).T))

def to_one_hot(x, d):
    o = torch.FloatTensor(x.shape[0], d).cuda()
    o.zero_()
    o.scatter_(1, x.unsqueeze(1), 1)
    return o

class Model(nn.Module):
    def __init__(self, d, hidden, num_inputs=2):
        super().__init__()
        self.m = {}
        self.d = d
        self.hidden = hidden
        self.mask1 = mask_operation(2*self.d, hidden[0])
        self.mask2 = mask_operation(hidden[0], hidden[1])
        self.mask3 = mask_operation(hidden[1], 2*self.d)
        self.relu = nn.ReLU()
        self.create()

    def forward(self, x):
        bs = x.shape[0]
        x = x.long().reshape(-1)
        x = to_one_hot(x, self.d)
        x = x.view(bs, -1)
        x = self.mask1(x)
        x = self.relu(x)
        x = self.mask2(x)
        x = self.relu(x)
        out = self.mask3(x)
        out = out.view(bs, 2, self.d)
        out = out.permute(0, 2, 1)
        return out

    def probs(self):
        x = np.mgrid[0:self.d, 0:self.d].reshape(2, self.d ** 2).T
        x = torch.LongTensor(x).cuda()
        log_probs = F.log_softmax(self(x), dim=1)
        distribution = torch.gather(log_probs, 1, x.unsqueeze(1)).squeeze(1)
        distribution = distribution.sum(dim=1)
        return distribution.exp().view(self.d, self.d).detach().cpu().numpy()

    def create(self):
        self.m[-1] = np.array([0, 1])
        for h in range(len(self.hidden)):
            self.m[h] = np.random.randint(self.m[h-1].min(), 1, size=self.hidden[ 0])
        masks = [self.m[h - 1][:, None] <= self.m[h][None, :] for h in range(len(self.hidden))]
        masks.append(self.m[len(self.hidden) - 1][:, None] < self.m[-1][None, :])
        masks[-1] = np.repeat(masks[-1], self.d, axis=1)
        masks[0] = np.repeat(masks[0], self.d, axis=0)
        layers = [l for l in self.modules() if isinstance(l, mask_operation)] 
        for l, m in zip(layers, masks):
            print(len(layers), len(masks))
            l.create(m)     

def train(model, optimizer, train_loader):
    model.train()
    loss_ = []
    for data in train_loader:
        optimizer.zero_grad()
        out = model(data.cuda().contiguous())
        loss = F.cross_entropy(out, data.cuda().long())
        loss_.append(loss.item())
        loss.backward()
        optimizer.step()
    return loss_

def eval(model, test_loader):
    model.eval()
    loss_ = 0
    with torch.no_grad():
        for data in test_loader:
            out = model(data.cuda().contiguous())
            loss = F.cross_entropy(out, data.cuda().long())
            loss_ += loss.item()
        return loss_/len(test_loader)   

def q2_a(train_data, test_data, d, dset_id):
    train_loader = DataLoader(train_data, shuffle=True, batch_size=128)
    test_loader = DataLoader(test_data, batch_size=128)
    model = Model(d=d, hidden=[100,100]).cuda()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.003)
    scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.1)
    tr_loss, te_loss = [], []
    te_loss.append(eval(model, test_loader))
    for i in range(10):
        tr_loss.extend(train(model, optimizer, train_loader))
        te_loss.append(eval(model, test_loader))
        scheduler.step()
    plt.plot(tr_loss)
    plt.show()
    plt.plot(te_loss)
    plt.show()
    probs = model.probs()
    return tr_loss, te_loss, probs

q2_save_results(1, 'a', q2_a)
```

Implement and train a MADE model on the given binary image datasets. Given some binary image of height $H$ and width $W$, we can represent image $x\in \{0, 1\}^{H\times W}$ as a flattened binary vector $x\in \{0, 1\}^{HW}$ to input into MADE to model $p_\theta(x) = \prod_{i=1}^{HW} p_\theta(x_i|x_{<i})$. Your model should output logits, after which you could apply a sigmoid over 1 logit, or a softmax over two logits (either is fine).

```python
#Finetuning necessary, sampling referred from original solution
class mask_operation(nn.Linear):
    def __init__(self, in_features, out_features, bias=True):
        super().__init__(in_features, out_features, bias)
        self.register_buffer('mask', torch.ones(out_features, in_features))
    
    def forward(self, x):
        return nn.functional.linear(x, self.mask*self.weight, self.bias)
  
    def create(self, mask):
        self.mask.data.copy_(torch.from_numpy(mask.astype(np.uint8).T))

def to_one_hot(x, d):
    o = torch.FloatTensor(x.shape[0], d).cuda()
    o.zero_()
    o.scatter_(1, x.unsqueeze(1), 1)
    return o

class Model(nn.Module):
    def __init__(self, n, hidden, input_, d=2):
        super().__init__()
        self.input_ = input_
        self.m = {}
        self.d = d
        self.n = n
        self.hidden = hidden
        self.mask1 = mask_operation(self.n*self.d, 512)
        self.mask2 = mask_operation(512, 512)
        self.mask3 = mask_operation(512, self.n*self.d)
        self.relu = nn.ReLU()
        self.create()
        self.order = np.arange(self.n)

    def forward(self, x):
        bs = x.shape[0]
        x = x.long().reshape(-1)
        x = to_one_hot(x, self.d)
        x = x.view(bs, -1)
        x = self.mask1(x)
        x = self.relu(x)
        x = self.mask2(x)
        x = self.relu(x)
        out = self.mask3(x)
        out = out.view(bs, self.n, self.d)
        out = out.permute(0, 2, 1).contiguous().view(bs, self.d, *self.input_)
        return out

    def sample100(self):
        samples = torch.zeros(100, self.n).cuda()
        with torch.no_grad():
            for i in range(self.n):
                logits = self(samples).view(100, self.d, self.n)[:, :, self.order[i]]
                probs = F.softmax(logits, dim=1)
                samples[:, self.order[i]] = torch.multinomial(probs, 1).squeeze(-1)
            samples = samples.view(100, *self.input_).cpu().numpy()
            samples = np.transpose(samples, (0, 2, 3, 1))
        return samples

    def create(self):
        self.m[-1] = np.arange(self.n)
        for h in range(len(self.hidden)):
            self.m[h] = np.random.randint(self.m[h-1].min(), self.n - 1, size=512)
        masks = [self.m[h - 1][:, None] <= self.m[h][None, :] for h in range(len(self.hidden))]
        masks.append(self.m[len(self.hidden) - 1][:, None] < self.m[-1][None, :])
        masks[-1] = np.repeat(masks[-1], self.d, axis=1)
        masks[0] = np.repeat(masks[0], self.d, axis=0)
        layers = [l for l in self.modules() if isinstance(l, mask_operation)] 
        for l, m in zip(layers, masks):
            l.create(m) 
            
def eval(model, test_loader):
    model.eval()
    loss_ = 0
    with torch.no_grad():
        for data in test_loader:
            out = model(data.cuda().contiguous())
            loss = F.cross_entropy(out, data.cuda().long())
            loss_ += loss.item()
        return loss_/len(test_loader) 

def q2_b(train_data, test_data, image_shape, dset_id):
    train_data = np.transpose(train_data, (0, 3, 1, 2))
    test_data = np.transpose(test_data, (0, 3, 1, 2))
    train_loader = DataLoader(train_data, shuffle=True, batch_size=128)
    test_loader = DataLoader(test_data, batch_size=128)
    model = Model(image_shape[0]*image_shape[1], input_ = (1, image_shape[0], image_shape[1]), hidden=[1024,1024]).cuda()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.003)
    scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.1)
    tr_loss, te_loss = [], []
    te_loss.append(eval(model, test_loader))
    for i in range(10):
        tr_loss.extend(train(model, optimizer, train_loader))
        te_loss.append(eval(model, test_loader))
        scheduler.step()
    plt.plot(tr_loss)
    plt.show()
    plt.plot(te_loss)
    plt.show()
    samples = model.sample100()

    return tr_loss, te_loss, samples

q2_save_results(1, 'b', q2_b)
```

## Part (a) PixelCNN on Shapes and MNIST
In this part, implement a simple PixelCNN architecture to model binary MNIST and shapes images (same as Q2(b), but with a PixelCNN).

We recommend the following network design:
* A $7 \times 7$ masked type A convolution

And the following hyperparameters:

Your model should output logits, after which you could apply a sigmoid over 1 logit, or a softmax over two logits (either is fine). It may also help to scale your input to $[-1, 1]$ before running it through the network. 

Training on the shapes dataset should be quick, and MNIST should take around 10 minutes

**You will provide these deliverables**

```python
#smaller model, colab keeps crashing(??)
def train(model, optimizer, train_loader, c):
    model.train()
    loss_ = []
    for data in train_loader:
        optimizer.zero_grad()
        if c:
            data = data.cuda()
        out = model(data.contiguous())
        loss = F.cross_entropy(out, data.long())
        loss_.append(loss.item())
        loss.backward()
        optimizer.step()
    return loss_

def eval_(model, test_loader):
    model.eval()
    loss_ = 0
    with torch.no_grad():
        for data in test_loader:
            data.cuda()
            out = model(data)
            loss = F.cross_entropy(out, data.cuda().long())
            loss_ += loss.item()
    return loss_/len(test_loader) 
import torch.nn as nn
from torch.utils import data
class mask(nn.Conv2d):
    def __init__(self, type_, no_channel, no_features, kernel_size, stride, padding=None):
        super().__init__(no_channel, no_features, kernel_size, stride, padding)
        self.kernel_size = kernel_size
        self.register_buffer('mask', torch.zeros_like(self.weight))
        self.create_mask(type_, self.kernel_size)

    def forward(self, x):
        return F.conv2d(x, self.weight * self.mask, self.bias, self.stride, self.padding, self.dilation, self.groups)
    
    def create_mask(self, type_, k):
        self.mask[:, :, :k // 2] = 1
        self.mask[:, :, k // 2, :k // 2] = 1
        if type_ == 'B':
            self.mask[:, :, k // 2, k // 2] = 1

class PixelCNN(nn.Module):
    def __init__(self, input_shape, no_channel=1, no_colours=2, no_features=64, kernel_size=7):
        super().__init__()
        self.input_shape = input_shape
        self.conv1 = mask('A', no_channel, no_features, kernel_size, 1, padding=int(kernel_size/2))
        self.conv2 = mask('A', no_features, no_features, kernel_size, 1, padding=int(kernel_size/2))
        self.conv3 = mask('A', no_features, no_features, kernel_size, 1, padding=int(kernel_size/2))
        self.conv4 = mask('B', no_features, no_colours, 1, 1, padding=0)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = (x.float()- 0.5) / 0.5
        x = self.relu(self.conv1(x.cuda()))
        x = self.relu(self.conv2(x))
        x = self.relu(self.conv3(x))
        x = self.conv4(x)
        return x.unsqueeze(dim=2)

    def sample100(self):
        samples = torch.zeros(100, *self.input_shape).cuda()
        with torch.no_grad():
            for r in range(self.input_shape[1]):
                for c in range(self.input_shape[2]):
                    logits = self(samples)[:, :, 0, r, c]
                    probs = F.softmax(logits, dim=1)
                    samples[:, 0, r, c] = torch.multinomial(probs, 1).squeeze(-1)
            return samples.permute(0, 2, 3, 1).cpu().numpy()
```
