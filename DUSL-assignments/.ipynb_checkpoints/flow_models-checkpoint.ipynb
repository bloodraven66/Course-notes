## Part (a) Autoregressive Flow [15pt]

In an autoregressive flow, we learn the transformation $z_i = f(x_i ; x_{1:i-1})$. The log-likelihood is then $\log p_\theta(x) = \sum_{i=1}^d \log p(z_i) + \log |\frac{dz_i}{dx_i}|$ because the Jacobian is triangular. For each dimension, use the CDF of a mixture of Gaussians or Logistics to map to the latent space, which should be $z_i \sim \text{Unif}[0, 1]$ i.i.d..

Fit $p_\theta$ with maximum likelihood via stochastic gradient descent on the training set. Since this is a 2D problem, you can either learn $z_0 = f(x_0)$ and $z_1 = f(x_1; x_0)$ together with a MADE model, or with separate networks. Use your favorite version of stochastic gradient descent, and optimize your hyperparameters on a validation set of your choice.

**You will provide these deliverables**


1.   Over the course of training, record the average negative log-likelihood (nats / dim) of the training data (per minibatch) and test data (for your entire test set). Code is provided that automatically plots the training curves. 
2.   Report the final test set performance of your final model
3. Show a heatmap of the densities learned by your model
4. Show what the training set looks like when mapped to the latent space. 




```python
import torch.nn as nn
import deepul.pytorch_util as ptu

import matplotlib.pyplot as plt
from torch.distributions.normal import Normal
from tqdm import tqdm_notebook as tqdm
def train(model, optimizer, train_loader):
    model.train()
    loss_ = []
    D = torch.distributions.uniform.Uniform(0, 1)
    for data in train_loader:
        optimizer.zero_grad()
        data = data.float().contiguous()
        out = model(data)
        loss = -(D.log_prob(out[0]) + out[1]).mean()
        loss_.append(loss.item())
        loss.backward()
        optimizer.step()
    return loss_

def eval_(model, test_loader):
    model.eval()
    loss_ = 0
    D = torch.distributions.uniform.Uniform(0, 1)
    with torch.no_grad():
        for data in test_loader:
            out = model(data.float())
            loss = -(D.log_prob(out[0]) + out[1]).mean()
            loss_ += loss.item()
    return loss_/len(test_loader) 


class MixOfGaussian(nn.Module):
    def __init__(self, no_components=4):
        super().__init__()
        self.no_components = no_components
        self.loc = nn.Parameter(torch.randn(self.no_components), requires_grad=True)
        self.log_scale = nn.Parameter(torch.zeros(self.no_components), requires_grad=True)
        self.theta = nn.Parameter(torch.zeros(self.no_components), requires_grad=True)
        
    def cdf(self, x):
        theta_w = F.softmax(self.theta, dim=0).unsqueeze(dim=0).repeat(x.shape[0], 1)
        D = Normal(self.loc, self.log_scale.exp())
        latent_z = (D.cdf(x.repeat(1, self.no_components))*theta_w).sum(dim=1)
        jacobian = (D.log_prob(x.repeat(1, self.no_components)).exp()*theta_w).sum(dim=1).log()
        return latent_z, jacobian
    
class MLP(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super().__init__()
        self.linear1 = nn.Linear(input_size, hidden_size)
        self.linear2 = nn.Linear(hidden_size, hidden_size)
        self.linear3 = nn.Linear(hidden_size, output_size)
        self.relu = nn.ReLU()
        
    def forward(self, x):
        x = self.relu(self.linear1(x))
        x = self.relu(self.linear2(x))
        return self.linear3(x)
        
class Model(nn.Module):
    def __init__(self):
        super().__init__()
        self.model1 = MixOfGaussian()
        self.model2 = MLP(1, 64, 3*4)
        self.no_components_row = 4
        self.no_components_col = 4
    
    def forward(self, x):
        x_row, x_col = x[:,0].unsqueeze(1), x[:,1].unsqueeze(1)
        latent_z1, jacobian1 = self.model1.cdf(x_row)
        out = self.model2(x_row)
        loc, log_scale, theta = torch.chunk(out, 3, 1)
        theta_w = F.softmax(theta)
        D = Normal(loc, log_scale.exp())
        latent_z2 = (D.cdf(x_col.repeat(1, self.no_components_col))*theta_w).sum(dim=1)
        jacobian2 = (D.log_prob(x_col.repeat(1, self.no_components_col)).exp()*theta_w).sum(dim=1).log()
        return torch.cat([latent_z1.unsqueeze(1), latent_z2.unsqueeze(1)], dim=1), torch.cat([jacobian1.unsqueeze(1), jacobian2.unsqueeze(1)], dim=1)
    
def q1_a(train_data, test_data, dset_id):
    D = torch.distributions.uniform.Uniform(0, 1)
    tr_loss, te_loss, densities, latent = [], [], [], []
    train_loader = torch.utils.data.DataLoader(train_data, batch_size=512, shuffle=True)
    test_loader = torch.utils.data.DataLoader(test_data, batch_size=512, shuffle=True)
    model = Model()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.005)
    te_loss.append(eval_(model, test_loader))
    for i in tqdm(range(1000)):
        tr_loss.extend(train(model, optimizer, train_loader))
        te_loss.append(eval_(model, test_loader))
    #given in code
    dx, dy = 0.025, 0.025
    if dset_id == 1:  
        x_lim = (-4, 4)
        y_lim = (-4, 4)
    elif dset_id == 2:  
        x_lim = (-1.5, 2.5)
        y_lim = (-1, 1.5)
    y, x = np.mgrid[slice(y_lim[0], y_lim[1] + dy, dy),slice(x_lim[0], x_lim[1] + dx, dx)]
    mesh_xs = torch.FloatTensor(np.stack([x, y], axis=2).reshape(-1, 2))
    latent_z, jacobian = model(mesh_xs)
    densities = np.exp((D.log_prob(latent_z).mean(dim=1) + jacobian.mean(dim=1)).detach().numpy())
    latent_z, _ = model(torch.FloatTensor(train_data))
    latents = latent_z.detach().detach().numpy()
    return tr_loss, te_loss, densities, latents
q1_save_results(1, 'a', q1_a)
```
